```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
```

```python
hf_tokenizer = Tokenizer(BPE())
hf_tokenizer.pre_tokenizer = Whitespace()
```

```python
trainer = BpeTrainer(vocab_size=16000, special_tokens=["<unk>"])

hf_tokenizer.train(["dialogues.txt"], trainer)
```

```python
hf_tokenizer.save("my_bpe_tokenizer_dia.json")
```

```python
!pip install datasets -q
from datasets import load_dataset
dataset = load_dataset("elricwan/dailydialog", data_dir="data")

print(dataset)
print(dataset["train"][0])
```

```python
train_pairs = []
for convo in dataset['train']:
    if len(convo['conversation']) >= 2:
        for i in range(len(convo['conversation']) - 1):
            train_pairs.append({
                'input': convo['conversation'][i],
                'output': convo['conversation'][i+1]
            })
```

```python
print(train_pairs[:2])
```

```python

num_examples = 10000
subset = train_pairs[:num_examples]

with open("dialogues.txt", "w", encoding="utf-8") as f:
    for item in subset:
        f.write(item['input'] + "\n")
        f.write(item['output'] + "\n")
        f.write("\n")

print(f"{len(subset)} çift diyalog 'dialogues.txt' dosyasına yazıldı.")
```
